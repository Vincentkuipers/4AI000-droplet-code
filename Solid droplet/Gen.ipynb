```python

edge_points = [0] * len(r_a)

verts = []
theta_step = 3.1415 / 16
for n in range(32):
    theta = n * theta_step
    print(theta)
    for i in range(len(r_a)):
        x = r_a[i]*np.sin(theta)
        
        r = r_a[i]*np.cos(theta)
        
        z = z_a[i]
        edge_points[i] = np.array((x,r,z))
    print(len(verts))
    verts.append(edge_points)
```
